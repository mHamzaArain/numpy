# Image Histograms with NumPy and Matplotlib

```python
img1 = plt.imread('/home/hamza/Desktop/LEARNING/Complete NumPy Masterclass Go from Zero to Hero in Numpy/21. Downloadable Code Bundle/CodeBundle/images/5.jpeg')

r = img1[:, :, 0]
g = img1[:, :, 1]
b = img1[:, :, 2]
```

```python
plt.subplots_adjust(hspace=0.5, wspace=0.5)

plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(img1)

hist, bins = np.histogram(r.ravel(), bins=256, range=(0, 256))
plt.subplot(2, 2, 2)
plt.title('Red Histogram')
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(g.ravel(), bins=256, range=(0, 256))
plt.subplot(2, 2, 3)
plt.title('Green Histogram')
plt.bar(bins[:-1], hist)

hist, bins = np.histogram(b.ravel(), bins=256, range=(0, 256))
plt.subplot(2, 2, 4)
plt.title('Blue Histogram')
plt.bar(bins[:-1], hist)

plt.show()
```
