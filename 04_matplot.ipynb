```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5) # [0 1 2 3 4]
y = x
plt.plot(x, y, 'o--')
plt.plot(x, -y, 'o-')
plt.title('y=x and y=-x')
plt.show()
```

```python
# Plot without axis
plt.plot(x, y, 'o--')
plt.axis('off')       # axis = 'off'
plt.show()
```

```python
x = np.arange(11) # [0 1 2 3 4]
y = np.linspace(0, 10, 11)
print(y)
plt.plot(x, y, 'o--')
plt.show()
```

```python
x = np.arange(11) # [0 1 2 3 4]
y = np.logspace(0.1, 1, 11)
print(y)
plt.plot(x, y, 'o--')
plt.show()
```

```python
# np.geimsapce() graph
y = np.geomspace(0.1, 1000, 11)
print(y)
plt.plot(x, y, 'o--')
plt.show()
```
